```python
fig = plt.figure(figsize=(15,25))
rows = 8
cols = 4

for i,col in enumerate(df.columns):
  ax = fig.add_subplot(rows, cols, i+1)

  ## Make the boxplot and the stripplot
  sns.stripplot(x = df[col], ax = ax, palette = 'vlag')
  sns.stripplot(x = df[col], ax = ax, size = 4, linewidth = 0)
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df.corr(),annot = True)
```

```python
pred = [col for col in df.columns if col != 'diagnosis']

fig = plt.figure(figsize = (18,25))
rows = 8
cols = 4

for i, col in enumerate(pred):
    ax = fig.add_subplot(rows, cols, i+1)
    sns.boxplot(data = df, x = 'diagnosis', y = col, ax = ax)
```

```python
X=df
X=X.drop(columns={'diagnosis'})
y=df['diagnosis']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
classifier = LogisticRegression()
parameters = {'penalty':['l1','l2','ElasticNet','None'],'C':[1,10,20,30,100]}
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(classifier,param_grid=parameters,cv=5)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
classifier = LogisticRegression(C=20,penalty='l2',solver = 'lbfgs')
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred,y_test))
```
