```python
df=df[['iyear','country_txt','region_txt','city','attacktype1_txt','targtype1_txt','gname','nkill','longitude']]
df.head()
```

```python
df=df.astype({'nkill':'int'})
```

```python

df['iyear'].value_counts(dropna = False).sort_index()
```

```python
#Number of Terrorist Activities each Year
import seaborn as sns
x_year = df['iyear'].unique()
y_count_years = df['iyear'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = x_year,
           y = y_count_years,
           palette = 'rocket')
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack_of_Years')
plt.show()
     
```

```python
#Terrorist Activities by Region in each Year through Area Plot
pd.crosstab(df.iyear,df.region_txt).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()
     
```

```python
df_region=df.region_txt.value_counts().to_frame().reset_index()
df_region.columns=['region_name','count']
```

```python
# total number of terrorist attacks in various regions of the world.
sns.barplot(y='region_name',x='count',data=df_region,orient='h',palette='flare',ec='black')
plt.grid(True)
plt.title("Number of Total Attacks in Each Region")
plt.ylabel('Region')
plt.xlabel('Number of attacks')
plt.show()
     
```

```python
#Top 10 countries with most attacks
top_10_country=df.country_txt.value_counts()[:10].to_frame().reset_index()
top_10_country.columns=['country_name','count']
```

```python

plt.figure(figsize=(15,5))
sns.barplot(data=top_10_country,x='country_name',y='count',ec='black',lw=1)
plt.grid(True)
plt.title('Top 10 Countries with most Attacks',fontsize=20)
plt.xlabel('Country',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

```python
#Target Methods by Terrorists
target_type=df.attacktype1_txt.value_counts()[0:10].to_frame().reset_index()
target_type.columns=['target_type','count']
```

```python

plt.figure(figsize=(25,10))
sns.barplot(x='target_type',y='count',data=target_type,palette='coolwarm',ec='black')
plt.xlabel('Target Type',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.title("Target Method by Terrorists",fontsize=30)
plt.grid(True)
```

```python
#City where terrorist killed most number of people in the world ?
top10_city=df.city.value_counts().to_frame()[0:10].reset_index()
top10_city.columns=['city','count']
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='city',y='count',data=top10_city,ec='black',palette='Set3')
plt.title("Most affected City in the World",fontsize=15)
plt.xlabel('Name of the City',fontsize=12)
plt.ylabel("Number of the Attacks",fontsize=12)
plt.grid(True)
plt.show()
```

```python
#Deadliest year in terms of death of people by terrorism
year=(df.groupby(['iyear'])['nkill'].size()).reset_index()
# sort the year dataset in descending order of the killed colum.
year=year.sort_values(by=['nkill'],ascending=False).head()
year
```

```python
plt.figure(figsize=(10,5))
color=['red','orange','pink','blue','yellow']
plt.bar(list(year['iyear']),list(year['nkill']),color=color,ec='black')
plt.grid(True)
plt.xlabel('Name of the Year')
plt.ylabel('Number of People Killed')
plt.show()
```
